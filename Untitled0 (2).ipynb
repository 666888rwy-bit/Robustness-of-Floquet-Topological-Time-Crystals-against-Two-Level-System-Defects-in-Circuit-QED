```python
import numpy as np
import matplotlib.pyplot as plt
from qutip import basis, sigmax, sigmam, mesolve, qeye, tensor, sigmaz, expect, rand_ket, ket2dm
```

```python

N = 6  # 先小一点
sx_list, sz_list = [], []

for i in range(N):
    op_list = [qeye(2)] * N
    op_list[i] = sigmax()
    sx_list.append(tensor(op_list))

    op_list[i] = sigmaz()
    sz_list.append(tensor(op_list))
```

```python
J = 1.0
h = 1.0

H1 = 0
for i in range(N-1):
    H1 += J * sz_list[i] * sz_list[i+1]

H2 = 0
for i in range(N):
    H2 += h * sx_list[i]
```

```python

T = 0.9*np.pi

U1 = (-1j * H1 * T/2).expm()
U2 = (-1j * H2 * T/2).expm()

UF = U2 * U1
```

```python
U1 = (-1j * H1 * T/2).expm()
U2 = (-1j * H2 * T/2).expm()
dt = T / 100
t_list = np.arange(0, 4*T, dt)
def H_t(t):
    if (t % T) < T/2:
        return H1
    else:
        return H2
```

```python
# Calculate eigenvalues and eigenvectors of the Floquet operator
evals, evecs = UF.eigenstates()
```

```python
psi0 = tensor([basis(2,0) for _ in range(N)])  # 全 ↑
n_steps = 50
psi = psi0

obs_edge = []
obs_bulk = []

for n in range(n_steps):
    obs_edge.append(expect(sz_list[0], psi))
    obs_bulk.append(expect(sz_list[N//2], psi))

    psi = UF * psi
```

```python
psi = psi0
data = np.zeros((N, len(t_list)))

for k, t in enumerate(t_list):
    for i in range(N):
        data[i, k] = expect(sz_list[i], psi)

    H = H_t(t)
    psi = ( (-1j * H * dt).expm() ) * psi
```

### FFT Analysis of Edge and Bulk Spin Expectation Values

To further understand the dynamics of the edge and bulk spins in the different Floquet phases, we will perform a Fast Fourier Transform (FFT) on the expectation values of $\langle\sigma^z\rangle$ for both the edge (first site) and bulk (middle site) over time. This analysis will reveal the dominant frequencies present in their oscillations and allow us to compare their spectral 'sharpness', which can indicate the coherence and stability of their dynamics.

```python
def plot_fft_comparison(ax, signal_edge, signal_bulk, title):
    # Remove DC component (mean subtraction)
    signal_edge_processed = signal_edge - np.mean(signal_edge)
    signal_bulk_processed = signal_bulk - np.mean(signal_bulk)

    # FFT for edge signal using rfft for real input
    fft_edge = np.fft.rfft(signal_edge_processed)
    power_edge = np.abs(fft_edge)**2
    freq_edge = np.fft.rfftfreq(len(signal_edge_processed), d=1) # rfftfreq returns non-negative frequencies

    # FFT for bulk signal using rfft for real input
    fft_bulk = np.fft.rfft(signal_bulk_processed)
    power_bulk = np.abs(fft_bulk)**2
    freq_bulk = np.fft.rfftfreq(len(signal_bulk_processed), d=1) # rfftfreq returns non-negative frequencies

    ax.plot(freq_edge, power_edge, label='Edge (FFT Power)', color='red')
    ax.plot(freq_bulk, power_bulk, label='Bulk (FFT Power)', linestyle='--', color='blue', alpha=0.7)
    ax.set_xlabel('Frequency (cycles per Floquet period)')
    ax.set_ylabel('Power (Amplitude$^2$)')
    ax.set_title(f'{title}')
    ax.grid(True)
    ax.legend()
    ax.set_xlim(0, 0.5) # Focus on the subharmonic range for time crystals
    ax.set_ylim(bottom=0) # Ensure y-axis starts from 0

# The previous calls to plot_fft_comparison will be removed and replaced by the new combined plotting cell.
# They are kept here for now as part of the modification step, but will effectively be superseded.
```

```python
fig, axes = plt.subplots(1, 3, figsize=(20, 6), sharey=True)

plot_fft_comparison(axes[0], np.array(obs_edge), np.array(obs_bulk), 'Original Initial State')
plot_fft_comparison(axes[1], np.array(obs_edge_rand), np.array(obs_bulk_rand), 'Random Initial State')
plot_fft_comparison(axes[2], np.array(obs_edge_alt), np.array(obs_bulk_alt), 'Alternating Initial State')

plt.suptitle('FFT Comparison of Edge and Bulk Spins for Different Initial States', fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # Adjust layout to prevent suptitle overlap
plt.show()
```

### Comparison: Open Boundary Conditions (OBC) vs. Periodic Boundary Conditions (PBC)

To strongly demonstrate edge localization, we will compare the Floquet dynamics under Open Boundary Conditions (OBC), which we have been using so far, with Periodic Boundary Conditions (PBC). Under PBC, the last site (`N-1`) interacts with the first site (`0`), effectively creating a ring. If edge modes are truly localized, their subharmonic ($f=0.5$) FFT peak should be significantly reduced or disappear under PBC, as the 'edge' behavior is smeared across the periodic chain.

```python
# --- Define Hamiltonians for Periodic Boundary Conditions (PBC) --- (N is 6, from prior setup)

# H1 for PBC: Ising interaction (J * sz_list[i] * sz_list[i+1]) plus periodic term
H1_pbc = 0
for i in range(N - 1):
    H1_pbc += J * sz_list[i] * sz_list[i+1]
H1_pbc += J * sz_list[N-1] * sz_list[0] # Periodic term

# H2 for PBC: Transverse field (h * sx_list[i]) - same as OBC as it applies to all sites
H2_pbc = 0
for i in range(N):
    H2_pbc += h * sx_list[i]

# --- Floquet operator for PBC ---
U1_pbc = (-1j * H1_pbc * T/2).expm()
U2_pbc = (-1j * H2_pbc * T/2).expm()
UF_pbc = U2_pbc * U1_pbc

print("PBC Hamiltonians and Floquet operator constructed.")
```

### Floquet Evolution for PBC (using the 'all-up' initial state)

Now, we run the Floquet evolution for the 'all-up' initial state under PBC and record the expectation values of the edge and bulk spins.

```python
psi0_pbc = tensor([basis(2,0) for _ in range(N)])  # All-up state
psi_pbc = psi0_pbc

obs_edge_pbc = []
obs_bulk_pbc = []

for n in range(n_steps):
    obs_edge_pbc.append(expect(sz_list[0], psi_pbc))
    obs_bulk_pbc.append(expect(sz_list[N//2], psi_pbc))

    psi_pbc = UF_pbc * psi_pbc

print("Floquet evolution for PBC complete.")
```

### FFT Comparison: OBC vs. PBC (for 'all-up' initial state)

### Quantitative Analysis: Ratio R = S_bulk(0.5) / S_edge(0.5)

To quantify the relative strength of the subharmonic spectral weight, we define $R = \frac{S_{bulk}(0.5)}{S_{edge}(0.5)}$, where $S(0.5)$ is the power at the $f=0.5$ frequency. A larger $R$ indicates stronger edge localization relative to the bulk.

```python
def get_power_at_nyquist(signal):
    signal_processed = signal - np.mean(signal)
    fft_result = np.fft.rfft(signal_processed)
    power_spectrum = np.abs(fft_result)**2
    freq = np.fft.rfftfreq(len(signal_processed), d=1)
    idx_0_5 = np.argmin(np.abs(freq - 0.5)) # Find index closest to 0.5
    return power_spectrum[idx_0_5]

# Calculate R for OBC (Original Initial State)
power_edge_obc_orig = get_power_at_nyquist(np.array(obs_edge))
power_bulk_obc_orig = get_power_at_nyquist(np.array(obs_bulk))
R_obc_orig = power_bulk_obc_orig / power_edge_obc_orig

# Calculate R for OBC (Random Initial State)
power_edge_obc_rand = get_power_at_nyquist(np.array(obs_edge_rand))
power_bulk_obc_rand = get_power_at_nyquist(np.array(obs_bulk_rand))
R_obc_rand = power_bulk_obc_rand / power_edge_obc_rand

# Calculate R for OBC (Alternating Initial State)
power_edge_obc_alt = get_power_at_nyquist(np.array(obs_edge_alt))
power_bulk_obc_alt = get_power_at_nyquist(np.array(obs_bulk_alt))
R_obc_alt = power_bulk_obc_alt / power_edge_obc_alt

# Calculate R for PBC (Original Initial State)
power_edge_pbc_orig = get_power_at_nyquist(np.array(obs_edge_pbc))
power_bulk_pbc_orig = get_power_at_nyquist(np.array(obs_bulk_pbc))
R_pbc_orig = power_bulk_pbc_orig / power_edge_pbc_orig

print(f"Quantitative Observable R (S_bulk(0.5) / S_edge(0.5)):")
print(f"  OBC - Original Initial State: R = {R_obc_orig:.4f}")
print(f"  OBC - Random Initial State:   R = {R_obc_rand:.4f}")
print(f"  OBC - Alternating Initial State: R = {R_obc_alt:.4f}")
print(f"  PBC - Original Initial State: R = {R_pbc_orig:.4f}")
```

### FFT Analysis of Edge Spin Expectation Values

To further understand the dynamics of the edge spins in the different Floquet phases, we will perform a Fast Fourier Transform (FFT) on the expectation values of $\langle\sigma^z_0\rangle$ (left edge site) over time for each scenario (MZM, MPM, and Anomalous Coexistence).

```python

import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import expm

N = 30
J = 1.0
h = 1.0

def run_continuous_floquet(T1, T2, periods=4, micro_steps=40):
    A_ZZ = np.zeros((2*N, 2*N))
    A_X  = np.zeros((2*N, 2*N))

    for i in range(N-1):
        A_ZZ[2*i+1, 2*i+2] = 2*J
        A_ZZ[2*i+2, 2*i+1] = -2*J
    for i in range(N):
        A_X[2*i, 2*i+1] = 2*h
        A_X[2*i+1, 2*i] = -2*h

    dt1 = T1 / micro_steps
    dt2 = T2 / micro_steps
    U_ZZ_step = expm(A_ZZ * dt1)
    U_X_step  = expm(A_X  * dt2)

    v = np.zeros(2*N)
    v[0] = 1.0   # 左边缘
    v[-1] = 1.0  # 右边缘

    data = []
    for p in range(periods):
        for _ in range(micro_steps):
            data.append(v[0::2] + v[1::2])
            v = U_ZZ_step @ v
        for _ in range(micro_steps):
            data.append(v[0::2] + v[1::2])
            v = U_X_step @ v

    data.append(v[0::2] + v[1::2])
    return np.array(data)

# ==========================================
# 🌌 严格根据文献相图配置参数 (The Magic Parameters)
# ==========================================
# 1. 纯 0-模相 (w0=1, w_pi=0): T1 占主导
data_MZM = run_continuous_floquet(T1=0.45*np.pi, T2=0.0)

# 2. 纯 pi-模相 (w0=0, w_pi=1): 完美的符号翻转
data_MPM = run_continuous_floquet(T1=0.45*np.pi, T2=0.9*np.pi)

# 3. 反常共存相 (w0=1, w_pi=1): PRR 2020 论文中的核心！
# 在这里，0模和pi模同时存在，波函数会自发产生完美的拍频 (Beating)
data_SUP = run_continuous_floquet(T1=0.75*np.pi, T2=0.75*np.pi)
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
titles = ['Isolated MZM\n($w_0=1, w_\pi=0$)',
          'Isolated MPM\n($w_0=0, w_\pi=1$)',
          'Anomalous Coexistence\n($w_0=1, w_\pi=1$)']
datas = [data_MZM, data_MPM, data_SUP]

# Y轴刻度对齐连续演化步数
micro_steps = 40
steps_per_period = micro_steps * 2
yticks = [0, steps_per_period, 2*steps_per_period, 3*steps_per_period, 4*steps_per_period]
yticklabels = ['0T', '1T', '2T', '3T', '4T']

for ax, data, title in zip(axes, datas, titles):
    # 【关键修复】加上 interpolation='bilinear' 让马赛克变成连续的光滑波纹
    # 使用 RdBu_r (红蓝配色) 更符合 PRR 论文中正负振幅交替的物理习惯
    im = ax.imshow(data, aspect='auto', origin='lower',
                   cmap='RdBu_r', vmin=-1.0, vmax=1.0,
                   interpolation='bilinear')
    ax.set_title(title, fontsize=14, pad=15, fontweight='bold')
    ax.set_xlabel('Site Index', fontsize=12)
    ax.set_ylabel('Time (Evolution $z$)', fontsize=12)
    ax.set_yticks(yticks)
    ax.set_yticklabels(yticklabels)

# 加个 Colorbar 显得更 Pro
cbar = fig.colorbar(im, ax=axes.ravel().tolist(), pad=0.02, aspect=40)
cbar.set_label('Wavefunction Amplitude $\psi$', fontsize=12)

plt.show()

```

```python
def plot_edge_fft(data, title):
    signal = data[:, 0]  # Left edge

    fft = np.fft.fft(signal)
    freq = np.fft.fftfreq(len(signal), d=1)  # Assuming a time step of 1 unit

    power = np.abs(fft)**2

    mask = freq >= 0  # Only consider positive frequencies

    plt.figure(figsize=(8, 5))
    plt.plot(freq[mask], power[mask])
    plt.xlabel("Frequency (1/period)")
    plt.ylabel("FFT Power")
    plt.title(title)
    plt.grid(True)
    plt.show()

# FFT for Isolated MZM
plot_edge_fft(data_MZM, 'FFT of Edge Spin for Isolated MZM')

# FFT for Isolated MPM
plot_edge_fft(data_MPM, 'FFT of Edge Spin for Isolated MPM')

# FFT for Anomalous Coexistence
plot_edge_fft(data_SUP, 'FFT of Edge Spin for Anomalous Coexistence')
```

```python
N = 30
J = 1.0
h = 1.0

A_ZZ = np.zeros((2*N, 2*N))
A_X  = np.zeros((2*N, 2*N))

for i in range(N-1):
    A_ZZ[2*i+1, 2*i+2] = 2*J
    A_ZZ[2*i+2, 2*i+1] = -2*J
for i in range(N):
    A_X[2*i, 2*i+1] = 2*h
    A_X[2*i+1, 2*i] = -2*h

T1 = 0.45*np.pi # T1 value taken from the MZM case in run_continuous_floquet call

T2_list = np.linspace(0, np.pi, 100)
all_phases = []
all_edge_weights = []

for T2 in T2_list:

    R_ZZ = expm(A_ZZ * T1)
    R_X  = expm(A_X  * T2)

    R_F = R_X @ R_ZZ   # 或试 R_ZZ @ R_X

    # Use np.linalg.eig to get both eigenvalues and eigenvectors
    evals, evecs = np.linalg.eig(R_F)
    phases = np.angle(evals)

    all_phases.append(phases)

    # Calculate edge weights for each eigenvector
    current_edge_weights = []
    for i in range(evecs.shape[1]): # Iterate over columns (eigenvectors)
        evec = evecs[:, i]
        # Edge sites are at 0, 1 (left) and -2, -1 (right) in the 2N dimension vector
        edge_weight = np.abs(evec[0])**2 + np.abs(evec[1])**2 + np.abs(evec[-2])**2 + np.abs(evec[-1])**2
        current_edge_weights.append(edge_weight)
    all_edge_weights.append(current_edge_weights)

all_phases = np.array(all_phases)  # shape: (len(T2_list), 2N)
all_edge_weights = np.array(all_edge_weights) # shape: (len(T2_list), 2N)

print("Floquet spectrum and edge weights calculated.")
```

```python
# Plot the Floquet spectrum, color-coded by edge weight
plt.figure(figsize=(10, 7))

# Flatten arrays for plotting
flat_T2_list = np.repeat(T2_list, all_phases.shape[1])
flat_phases = all_phases.flatten()
flat_edge_weights = np.array(all_edge_weights).flatten()

# Create a scatter plot where color represents the edge weight
# Using a diverging colormap like 'coolwarm' or 'RdBu_r' can be effective
# For edge weights, 'viridis' or 'plasma' could also work, or a simple 'hot' for high values.
# Let's use 'viridis' for now, where higher values mean more edge localization
scatter = plt.scatter(flat_T2_list, flat_phases, c=flat_edge_weights, cmap='viridis', s=10, alpha=0.7)

plt.xlabel("T2")
plt.ylabel("Quasienergy (phase)")
plt.title("Floquet Spectrum Color-Coded by Edge Weight")
plt.ylim([-np.pi, np.pi])
plt.colorbar(scatter, label='Edge Weight (abs(psi_left)^2 + abs(psi_right)^2)')
plt.grid(True)
plt.show()
```

```python


N = 6                # 时间晶体主链的比特数
N_total = N + 1      # 总比特数 (主链 + 1个TLS缺陷)

J = 1.0              # 主链内部的 Ising 相互作用强度
h = 1.0              # 主链的横场驱动强度

# Floquet 脉冲时间 (稍微偏离完美点，展示拓扑保护)
T1 = 0.85 * (np.pi / 4)
T2 = 0.95 * (np.pi / 2)

# TLS 缺陷的关键参数 (你的 Proposal 核心变量！)
g_defect = 0.5       # TLS 与边缘比特的耦合强度 (你可以试着改成 0.1 或 2.0 看看区别)
gamma_TLS = 0.01      # TLS 向环境漏能量的耗散率 (Purcell decay)

periods = 40         # 我们要模拟的 Floquet 周期数
```

```python
from qutip import destroy # Add this import statement

sz_list = []
sx_list = []
sm_list = [] # 降算符 (sigma_-)，用于写能量交换和耗散

for i in range(N_total):
    op_list = [qeye(2)] * N_total

    op_list[i] = sigmaz()
    sz_list.append(tensor(op_list))

    op_list[i] = sigmax()
    sx_list.append(tensor(op_list))

    op_list[i] = destroy(2) # QuTiP 里的 sigma_-
    sm_list.append(tensor(op_list))

print(f"成功构建 {N_total} 体系张量空间！矩阵维度: {sz_list[0].shape}")
```

```python

# H1: Ising 相互作用 (ZZ项)
H1 = 0
for i in range(N_total - 2):  # 主链的 N-1 个连接
    H1 += J * sz_list[i] * sz_list[i+1]
H1 += g_defect * sx_list[N-1] * sz_list[N_total-1] # TLS 和最右边的比特耦合

# H2: 横场驱动 (X项)
H2 = 0
for i in range(N_total - 1): # 对所有主链比特施加横场
    H2 += h * sx_list[i]

print(f"H1 维度: {H1.dims}, H2 维度: {H2.dims}")
```

```python
# 两个 Floquet 脉冲的演化算符
U1 = (-1j * H1 * T1).expm()
U2 = (-1j * H2 * T2).expm()

# 组合成一个完整的 Floquet 演化算符
UF = U2 * U1

print(f"U1 维度: {U1.dims}, U2 维度: {U2.dims}, UF 维度: {UF.dims}")
```

```python
# 1. 主链内部相互作用 (ZZ)
H_ZZ = 0
for i in range(N-1):
    H_ZZ += -J * sz_list[i] * sz_list[i+1]

# 2. 主链外部驱动 (X) - 注意只作用于主链 (range N)，不包括 TLS！
H_X = 0
for i in range(N):
    H_X += -h * sx_list[i]

# 3. TLS 幽灵缺陷的耦合 (Jaynes-Cummings 能量交换形式)
# TLS (索引 N) 只和左边缘比特 (索引 0) 耦合
H_defect = g_defect * (sm_list[0].dag() * sm_list[N] + sm_list[0] * sm_list[N].dag())

# 拼装 Floquet 两步哈密顿量
H1 = H_ZZ + H_defect  # 第一步：内部相互作用 + 缺陷始终在吸血
H2 = H_X + H_defect   # 第二步：外部横场翻转 + 缺陷始终在吸血

# 4. 定义开放量子系统的耗散通道 (Lindblad Collapse Operators)
# TLS 正在向环境漏气
c_ops = [np.sqrt(gamma_TLS) * sm_list[N]]

print("哈密顿量与耗散通道构建完毕！")
```

```python
# 初始化：边缘比特设为激发态(Up)，其余所有比特(包括体态和TLS)都在基态(Down)
# QuTiP中 basis(2,0)是态|0> (Z期望值为+1), basis(2,1)是态|1> (Z期望值为-1)
state_list = [basis(2,0)] + [basis(2,1)] * (N-1) + [basis(2,1)]
psi0 = tensor(state_list)
rho0 = ket2dm(psi0) # 转换为密度矩阵

# 用于存储期望值
expect_edge = []
expect_bulk = []
expect_tls  = []

rho_t = rho0
print("开始进行开放量子系统 Floquet 演化...")

for p in range(periods):
    # 记录当前周期的可观测量
    expect_edge.append(expect(sz_list[0], rho_t))       # 边缘态
    expect_bulk.append(expect(sz_list[N//2], rho_t))    # 体态
    expect_tls.append(expect(sz_list[N], rho_t))        # TLS 幽灵缺陷

    result1 = mesolve(H1, rho_t, [0, T1], c_ops, [])
    rho_mid = result1.states[-1]

    result2 = mesolve(H2, rho_mid, [0, T2], c_ops, [])
    rho_t = result2.states[-1]

print("演化完成！")
```

### FFT Spectrum: Frequency Domain Evidence of Time Crystal

To observe the "frequency domain evidence" of the time crystal, we perform a Fast Fourier Transform (FFT) on the expectation value of the edge spin ($\langle\sigma^z_0\rangle$) over time. A distinct peak at $\omega/2$ (subharmonic) in the frequency spectrum indicates the breaking of discrete time-translation symmetry and the presence of a time crystal. Conversely, a broad, noisy spectrum suggests the collapse of the time crystal due to dissipation.

```python
import numpy as np
import matplotlib.pyplot as plt

# Extract the signal from expect_edge
signal = np.array(expect_edge)

# Perform FFT
fft_result = np.fft.fft(signal)

# Calculate frequencies
freq = np.fft.fftfreq(len(signal), d=1) # d=1 implies a sampling interval of 1 (one Floquet period)

# Calculate power spectrum (amplitude squared)
power_spectrum = np.abs(fft_result)**2

# Filter for positive frequencies only
positive_freq_mask = freq >= 0

plt.figure(figsize=(10, 6))
plt.plot(freq[positive_freq_mask], power_spectrum[positive_freq_mask])
plt.xlabel('Frequency ($\\omega / 2\\pi$ per Floquet Period)')
plt.ylabel('Power (Amplitude$^2$)')
plt.title('FFT Spectrum of Edge Spin Expectation Value')
plt.grid(True)
plt.xlim(0, 0.5) # Focus on the subharmonic peak which should be at 0.5 for period-doubling
plt.show()
```

### Phase Diagram Heatmap: Mapping Time Crystal Survival

Next, we will generate a "Phase Diagram Heatmap" by scanning two critical parameters: `g_defect` (TLS-edge coupling strength) and `gamma_TLS` (TLS dissipation rate). For each combination, we will run the Floquet simulation and quantify the "survival" of the time crystal by measuring the average oscillation amplitude of the edge qubit in the final few periods. This heatmap will reveal the phase boundary between the robust topological phase (time crystal survival) and the dissipative phase (time crystal collapse), offering insights into the fault tolerance of the system.

```python

g_defect_list = np.linspace(0, 2.0, 20) # 20 points from 0 to 2.0
gamma_TLS_list = np.linspace(0, 1.0, 20) # 20 points from 0 to 1.0

oscillation_amplitude_heatmap = np.zeros((len(gamma_TLS_list), len(g_defect_list)))


print("Starting phase diagram calculation...")

for i, g_defect_val in enumerate(g_defect_list):
    for j, gamma_TLS_val in enumerate(gamma_TLS_list):
        # Update g_defect and gamma_TLS for the current iteration
        g_defect = g_defect_val
        gamma_TLS = gamma_TLS_val

        H_defect = g_defect * (sm_list[0].dag() * sm_list[N] + sm_list[0] * sm_list[N].dag())


        H_ZZ_const = 0
        for k_idx in range(N-1):
            H_ZZ_const += -J * sz_list[k_idx] * sz_list[k_idx+1]

        H_X_const = 0
        for k_idx in range(N):
            H_X_const += -h * sx_list[k_idx]

        H1_current = H_ZZ_const + H_defect  # First step: internal interaction + defect
        H2_current = H_X_const + H_defect   # Second step: external field + defect

        # Lindblad Collapse Operators
        c_ops_current = [np.sqrt(gamma_TLS) * sm_list[N]]

        # --- Re-run the open system Floquet evolution --- (similar to cell B493ZtDjtlHB)
        # Initialization (psi0/rho0 are already defined earlier in cell B493ZtDjtlHB and global)
        # However, for a clean sweep, let's ensure rho0 is always fresh.
        state_list_reinit = [basis(2,0)] + [basis(2,1)] * (N-1) + [basis(2,1)] # Ensure N is correct
        psi0_reinit = tensor(state_list_reinit)
        rho0_reinit = ket2dm(psi0_reinit)

        expect_edge_current = []
        rho_t_current = rho0_reinit

        for p in range(periods):
            expect_edge_current.append(expect(sz_list[0], rho_t_current))

            result1 = mesolve(H1_current, rho_t_current, [0, T1], c_ops_current, [])
            rho_mid = result1.states[-1]

            result2 = mesolve(H2_current, rho_mid, [0, T2], c_ops_current, [])
            rho_t_current = result2.states[-1]

        # Calculate "average oscillation amplitude" from the last 5 periods
        # A simple metric: max value - min value in the last few periods
        if len(expect_edge_current) >= 5:
            last_5_periods_data = expect_edge_current[-5:]
            amplitude = np.max(last_5_periods_data) - np.min(last_5_periods_data)
        else:
            # Handle cases where periods < 5 (should not happen with periods=40)
            amplitude = 0.0 # Or some other default value

        oscillation_amplitude_heatmap[j, i] = amplitude

print("Phase diagram calculation complete. Plotting heatmap...")

# Plotting the heatmap
plt.figure(figsize=(12, 8))
plt.pcolormesh(g_defect_list, gamma_TLS_list, oscillation_amplitude_heatmap, shading='auto', cmap='RdBu_r', vmin=0, vmax=2) # vmax=2 because sigma_z ranges from -1 to 1
plt.colorbar(label='Avg. Oscillation Amplitude (last 5 periods)')
plt.xlabel('g_defect (TLS-Edge Coupling Strength)')
plt.ylabel('gamma_TLS (TLS Dissipation Rate)')
plt.title('Phase Diagram: Edge Spin Oscillation Amplitude')
plt.show()
```
